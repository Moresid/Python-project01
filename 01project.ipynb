```python
yaricap = int(input("dairenin yarı çapını giriniz: "))

alan= 3*yaricap*yaricap

cevre= 2*3*yaricap

print(50*"*")

print(f"Pi: {3} \nYarıçap: {yaricap}, \nAlan: {alan}, \nÇevre: {cevre}")
```

```python
akenar= int(input("A kenarını giriniz: "))
bkenar= int(input("B kenarını giriniz: "))
ckenar= int(input("C kenarını giriniz: "))

alan= 2*(akenar*bkenar+bkenar*ckenar+akenar*ckenar)

hacim= akenar*bkenar*ckenar

print(50*"*")

print(f"A kenarı: {akenar}, B kenarı: {bkenar}, C kenarı: {ckenar} \n Alan: {alan}², \n Hacim: {hacim}³")
```
